```python
import math
```

```python
#sigmoid
#1/(1+e^-z)
#returns anything between 0 and 1

def sigmoid(x):
  return (1/(1+math.exp(-x)))
```

```python
print(sigmoid(100))
print(sigmoid(2))
print(sigmoid(-100))
```

```python
#tanh function
#(e^z-e^-z)/e^z+e^-z)
#returns anything between 1 and -1
def tanh(x):
  return ((math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x)))
```

# tanh.png

```python
print(tanh(100))
print(tanh(2))
print(tanh(-100))
```

```python
# relu function
# returns max of 0 or value
def relu(x):
  return (max(0,x))
```

```python
print(relu(100))
print(relu(2))
print(relu(-100))
```

```python
# relu function
# returns max of 0 or value
def leky_relu(x):
  return (max(0.1*x,x))
```

```python
print(leky_relu(100))
print(leky_relu(2))
print(leky_relu(-100))
```
